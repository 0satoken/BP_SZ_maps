# Loading an all-sky map with halo-based pasting

First, we import modules for loading maps. The all-sky map is healpix format. The healpy package is required.

```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
%matplotlib inline
```

Then, we load tSZ and kSZ fits files.

```python
ymap = hp.read_map("allsky_HP_tSZ_nres13_r000.fits")
bmap = hp.read_map("allsky_HP_kSZ_nres13_r000.fits")

print("y map Nside:", hp.get_nside(ymap))
print("b map Nside:", hp.get_nside(bmap))
```

We plot tSZ and kSZ in mollewide view.

```python
hp.mollview(ymap, norm='hist', title="All-sky HP tSZ")
```

```python
hp.mollview(bmap, norm='hist', title="All-sky HP kSZ")
```
